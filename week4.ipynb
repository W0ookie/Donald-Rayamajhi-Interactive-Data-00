```python
eval_counts = df["Evaluation"].value_counts()
```

```python
fig, ax = plot.subplots()

ax.bar(eval_counts.index.astype(str), eval_counts.values)
ax.set_title("Frequencies of Evaluation Responses")
ax.set_xlabel("Evaluation")
ax.set_ylabel("Count")

plot.show()
```

```python
fig, ax = plot.subplots()

ax.scatter(df["ConMediaUse"], df["ImmiPhobia"], alpha=0.6)

ax.set_title("ConMediaUse vs ImmiPhobia")
ax.set_xlabel("ConMediaUse")
ax.set_ylabel("ImmiPhobia")

plot.show()
```
